# Examining Linearity of Planck function at 4$\mu$m

```python
import numpy as np

from plumerise.planck import B21

import matplotlib.pyplot as plt
%matplotlib inline  
```

## How linear is the Planck function at 4$\mu$m as a function of $T^4$

```python
T = np.arange(300,2000)
T4 = T**4
P = B21(T)
slope = (P[500]-P[0])/(T4[500]-T4[0])
PL = P[0] + slope * (T4-T4[0])
slope
         
```

```python
plt.plot(T**4,P,label='nonlinear')
plt.plot(T**4,PL,'r--',label='nonlinear')
plt.xlabel(r'$T^4$'), plt.ylabel(r'$P(T^4)$'), plt.title(r'How Linear is $B_{4\mu m}$(T^4)')
plt.grid(), plt.legend()
```

```python
plt.plot(T,P,label='nonlinear')
plt.plot(T,PL,'r--',label='nonlinear')
plt.xlabel(r'$T$'), plt.ylabel(r'$P(T^4)$'), plt.title(r'How Linear is $B_{4\mu m}$(T^4)')
plt.grid(), plt.legend()
```
