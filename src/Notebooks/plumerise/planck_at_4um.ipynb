# Examining Linearity of Planck function at 4$\mu$m

```python
import numpy as np

from plumerise.planck import B21, iB21

import matplotlib.pyplot as plt
%matplotlib inline  
```

## How linear is the Planck function at 4$\mu$m as a function of $T^4$

```python
T = np.arange(300,2000)
T4 = T**4
P = B21(T)
slope = (P[500]-P[0])/(T4[500]-T4[0]) # based on 800K and 300K.
PL = P[0] + slope * (T4-T4[0])
slope
         
```

```python
plt.plot(T**4,P,label='nonlinear')
plt.plot(T**4,PL,'r--',label='nonlinear')
plt.xlabel(r'$T^4$'), plt.ylabel(r'$B_{4\mu m}(T^4)$'), plt.title(r'How Linear is $B_{4\mu m}(T^4)$')
plt.grid(), plt.legend()
```

```python
plt.plot(T[0:500]**4,P[0:500],label='nonlinear')
plt.plot(T[0:500]**4,PL[0:500],'r--',label='nonlinear')
plt.xlabel(r'$T^4$'), plt.ylabel(r'$B_{4\mu m}(T^4)$'), plt.title(r'How Linear is $B_{4\mu m}(T^4)$')
plt.grid(), plt.legend()
```

```python
plt.plot(T,P,label='nonlinear')
plt.plot(T,PL,'r--',label='nonlinear')
plt.xlabel(r'$T$'), plt.ylabel(r'$B_{4\mu m}(T)$'), plt.title(r'How Linear is $B_{4\mu m}$(T^4)')
plt.grid(), plt.legend()
```

## Brightness temperature rectification
The resulting modified BT is proportional to $T^4$.

```python
factor = PL/P
plt.plot(P,factor)
plt.xlabel(r'$B_{4\mu m}$'), plt.ylabel('Correction Factor'), plt.title('Correction Factor: Linear/Nonlinear')
plt.grid()
```

```python
factor = PL/P
plt.plot(T,factor)
plt.xlabel(r'$T$'), plt.ylabel('Correction Factor'), plt.title('Correction Factor: Linear/Nonlinear')
plt.grid()
```

## A Simpler Solution Using the Inverse of the Planck Function
A simpler solution is to simply calculate the inverse of the Planck function and take the 4th power. The empirical regression should be computed in terms of
$$
      \widetilde{T^4} = \left [  B_{4\mu m}^{-1} \left (\Delta L_{4\mu m}(T) \right )\right ]^4
$$
where
$$
\qquad \Delta L_{4\mu m}(T) = L_{4\mu m}(T)- L_{4\mu m}(T_B)
$$
Of course, one need to investigate the impact of non-homogeneity, in particular in the presence of flaming and smoldering components. By construction, this is on the 1:1 line.

```python
iT = iB21(P) # Brightness Temperature
E = iT**4
plt.plot(T4,E)
plt.xlabel(r'$T^4$'), plt.ylabel(r'$\left [B_{4\mu m}^{-1} \right ]^4$'), plt.title('Trivial Correction')
plt.grid()
```
